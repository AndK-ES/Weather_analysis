## Z-score

```python
def find_anomalies(series):
    Q1 = series.quantile(0.25)
    Q3 = series.quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    return (series < lower_bound) | (series > upper_bound)

df['is_anomaly'] = df.groupby('month')['t average, C'].transform(find_anomalies)
anomalies = df[df['is_anomaly']]
anomalies
```

```python
plt.figure(figsize=(15,5))
plt.plot(df.index, df['t average, C'], label = 'temp')
plt.scatter(anomalies.index, anomalies['t average, C'], color = 'red', label = 'anomalies')
plt.xlabel('date')
plt.ylabel('temp, C')
plt.legend()
plt.savefig(os.path.join('../figs', 'z-scores_t-avg.png'))
plt.show()
```

```python
winter_months = [12,1,2]
summer_months = [6,7,8]

winter_anomalies = df[df['is_anomaly'] & df['month'].isin(winter_months)]
summer_anomalies = df[df['is_anomaly'] & df['month'].isin(summer_months)]

plt.figure(figsize=(15,10))
plt.plot(df.index, df['t average, C'], label = 'temperature')
plt.scatter(winter_anomalies.index, winter_anomalies['t average, C'], color = 'blue', label = 'winter anomaly')
plt.scatter(summer_anomalies.index, summer_anomalies['t average, C'], color = 'red', label = 'summer_anomaly')
plt.legend()
plt.savefig(os.path.join('../figs', 'z-scores_t-winter-summer.png'))
plt.show()
```

## Extreme values

```python
max_temp = df['t average, C'].max()
max_date = df['t average, C'].idxmax()
min_temp = df['t average, C'].min()
min_date = df['t average, C'].idxmin()

print(f'Maximum of the years 2010-2020: {max_temp} Â°C, {max_date}')
print(f'Minimum of the years 2010-2020: {min_temp} Â°C, {min_date}')
```

```python
# local extreme values

from scipy.signal import argrelextrema

local_max_idx = argrelextrema(df['t average, C'].values, np.greater)[0]
local_min_idx = argrelextrema(df['t average, C'].values, np.less)[0]

local_max = df.iloc[local_max_idx]
local_min = df.iloc[local_min_idx]
```

```python
plt.figure(figsize=(70,20))
plt.plot(df.index, df['t average, C'], label = 'temperature')
plt.scatter(local_max.index, local_max['t average, C'], label = 'local max extreme', color = 'red')
plt.scatter(local_min.index, local_min['t average, C'], label = 'local min extreme', color = 'blue')
plt.legend()
plt.savefig(os.path.join('../figs', 'local_min-max_extremes.png'))
plt.show()

```
