```python
import pandas as pd
from datetime import datetime
from meteostat import Point, Daily
```

```python
madrid = Point(40.4168, -3.7038, 667)
```

```python
start = datetime(2010, 1, 1)
end = datetime(2020, 12, 31)
```

```python
data = Daily(madrid, start, end)
data = data.fetch()
data
```

```python
data.to_csv("../data/raw/madrid_weather_2010-2022.csv", index=True)
```
