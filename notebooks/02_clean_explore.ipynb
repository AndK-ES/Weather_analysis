# 02 - Data Cleaning and Exploration

```python
import pandas as pd
import numpy as np
from datetime import datetime
```

```python
df = pd.read_csv('../data/raw/madrid_weather_2010-2022.csv')
df
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
df = df.dropna(axis = 1, how = 'all').fillna(0)
df
```

```python
df = df.rename(columns = {'tavg' : 't average, C', 'tmin' : 't minimum, C', 'tmax' : 't maximum, C' , 
              'prcp' : 'precipitation, mm', 'wspd' : 'wind speed, km/h', 'pres': 'pressure, Pa'})
df
```

```python
df = df.set_index('time')
df.index = pd.to_datetime(df.index)
df
```

```python
df.to_csv('../data/clean/madrid_weather_clean.csv')
```

```python
df['year'] = df.index.year
df['month'] = df.index.month
df['day_of_year'] = df.index.dayofyear
df['day_of_week'] = df.index.dayofweek
df
```

```python
df['range'] = df['t maximum, C'] - df['t minimum, C']
df
```

```python
df['t_avg_7days'] = df['t average, C'].rolling(window = 7).mean()
df['t_avg_30days'] = df['t average, C'].rolling(window = 30).mean()
df
```
